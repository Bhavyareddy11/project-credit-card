```r
#the 4 strong corelation variables are v2,v5,v7,v20
cor(cc$Amount,cc$V1)
cor(cc$Amount,cc$V2)
cor(cc$Amount,cc$V5)
cor(cc$Amount,cc$V6)
cor(cc$Amount,cc$V7)
cor(cc$Amount,cc$V20)

```

```r
#lineatregration for significant variables
linearmodel <- lm(cc$Amount ~ cc$V1+cc$V2+cc$V5+cc$V6+cc$V7+cc$V20, data=cc)
summary(linearmodel)

```

```r
#linearregression model 
#creating training and test data
set.seed(100)
trainingRowIndex <- sample(1:nrow(cc), 0.8*nrow(cc))  # row indices for training data
trainingData <- cc[trainingRowIndex, ]  # model training data
testData  <- cc[-trainingRowIndex, ]   # test data
```

```r
#testData
```

```r
nrow(testData)
```

```r
nrow(trainingData)
```

```r
#Build the training data
linearmode2 <- lm(Amount ~ V1+V2+V5+V6+V7+V20, data = trainingData)
linearmode2
```

```r
#testing the train data
pred <- predict(linearmode2 ,testData )
head(pred)
```

```r
#diff btw the prediction and test data
result <- data.frame(Amount = testData$Amount , prediction = pred, diff = (testData$Amount - pred))
result
```

```r
#plotting amount ~ impvariable
par(mfrow = c(2,2))
plot(testData$Amount ~ testData$V1+testData$V2+testData$V5+testData$V6+testData$V7+testData$V20 , col = "red" , main = "regression" , data = testData)
abline(lm(testData$Amount ~ testData$V1+testData$V2+testData$V5+testData$V6+testData$V7+testData$V20 ,  data=testData))
```
